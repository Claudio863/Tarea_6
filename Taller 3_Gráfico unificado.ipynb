```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pwd='/home/mgcl/Escritorio/Quinto semestre/Taller DCD 3/Tareas/Tarea 3/for007.dat'
data = np.loadtxt(pwd)
data_a=np.array(data)
```

```python
df = pd.DataFrame(data_a, columns=['col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9', 'col10', 'col11', 'col12', 'col13', 'col14', 'col15', 'col16','col17'])
#print(df['col1'])
```

```python
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(20, 23))

#Grfico 1
axs[0,0].plot(df['col2'], df['col3'], 'b', label='HYDWIND', linewidth=2)
axs[0,0].set_xlabel('col 2', fontsize=12)
axs[0,0].set_ylabel('col 3', fontsize=12)
axs[0,0].set_yticks(range(0, 451, 50))
axs[0,0].set_title('Velocity profile', fontsize=20)
axs[0,0].grid(True)
axs[0,0].legend(loc='upper left')

#Grfico 2
axs[0,1].plot(df['col1'], df['col3'],'b',linewidth=2)
axs[0,1].set_xlim(1, 1.04)
axs[0,1].set_ylim(0, 20)
axs[0,1].set_xticks(np.arange(1, 1.040, 0.005))
axs[0,1].set_yticks(np.arange(0, 20.1, 2.5))
axs[0,1].set_xlabel('Col 1',fontsize=12)
axs[0,1].set_ylabel('Col 3',fontsize=12)
axs[0,1].set_title('V(r/R*) (inner)',fontsize=20)
axs[0,1].grid(True)

#Grfico 3
axs[1,0].plot(df['col2'], df['col17'], 'r', label='TRIAL PROFILE', linewidth=2)
axs[1,0].plot(df['col2'],df['col3'], 'b',label='HYDWIND',linewidth=2)
axs[1,0].set_xlabel('col 2', fontsize=12)
axs[1,0].set_ylabel('col 17', fontsize=12)
axs[1,0].set_yticks(range(0, 451, 50))
axs[1,0].set_title('Velocity profile', fontsize=20)
axs[1,0].grid(True)
axs[1,0].legend(loc='upper left')

#Grfico 4
axs[1,1].plot(df['col1'], df['col3'],'b',linewidth=2)
axs[1,1].set_xlim(70, 100)
axs[1,1].set_ylim(380, 420)
axs[1,1].set_xticks(np.arange(70, 101, 5))
axs[1,1].set_yticks(np.arange(380, 421, 5))
axs[1,1].set_xlabel('Col 1',fontsize=12)
axs[1,1].set_ylabel('Col 3',fontsize=12)
axs[1,1].set_title('V(r/R*) (outer)',fontsize=20)
axs[1,1].grid(True)

#Grfico 5
axs[2, 0].plot(df['col2'], df['col4'], 'b', linewidth=2)
axs[2, 0].set_xlabel('col 2', fontsize=12)
axs[2, 0].set_ylabel('col 4', fontsize=12)
axs[2, 0].set_yticks(range(0, 121, 20))
axs[2, 0].set_title('dw/du', fontsize=20)
axs[2, 0].grid(True)

#Grafico 6
x = np.log(df['col1'] - 1.0)
y = np.log(df['col6'])
axs[2, 1].plot(x, y, 'b', linewidth=2)
axs[2, 1].set_xlabel('log(col 1 - 1.0)', fontsize=12)
axs[2, 1].set_ylabel('log(col 6)', fontsize=12)
axs[2, 1].set_title('Density', fontsize=20)
axs[2, 1].grid(True)

#Grfico 7
axs[3,0].plot(df['col1'], df['col9'],'b',linewidth=2)
axs[3,0].set_xlabel('col 1',fontsize=12)
axs[3,0].set_ylabel('col 9',fontsize=12)
axs[3,0].set_title('g_line',fontsize=20)
axs[3,0].set_xlim(0.0, 20.0)
axs[3,0].set_ylim(0, 1600)
axs[3,0].grid(True)

# Oculta el ltimo grfico vaco
axs[-1, -1].axis('off')


plt.tight_layout()
plt.show()
```
