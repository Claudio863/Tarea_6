Importar las siguientes librerías: `matplotlib.pyplot, seaborn, pandas, numpy`

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## <b>Recrea la siguiente figura</b>

```python

# Generar datos para las distribuciones normales
mu1 = 0  # Media para la primera distribución
mu2 = 0  # Media para la segunda distribución
sigma1 = 1.5  # Desviación estándar para la primera distribución
sigma2 = 1.7  # Desviación estándar para la segunda distribución
size = 1000  # Tamaño de las muestras

np.random.seed(0)
x = np.linspace(-4, 4, size)
y1 = (1 / (sigma1 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu1) / sigma1) ** 2)
y2 = (1 / (sigma2 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu2) / sigma2) ** 2)

# Crear la figura y los subplots
fig, ax = plt.subplots(figsize=(10, 8))

# Gráfico de las distribuciones normales
ax.plot(x, y1, label='Distribución 1',linestyle='-.',linewidth=5,color='r')
ax.plot(x, y2, label='Distribución 2',linestyle='dashed',marker='p',markersize=4,color='yellow')

# Gráfico pequeño dentro del gráfico de las distribuciones
ax_inset = ax.inset_axes([0.35, 0.35, 0.3, 0.3])
ax_inset.text(0.5, 0.5, 'Cool plot bro', transform=ax_inset.transAxes, fontsize=12, ha='center', va='center', rotation=-45)
ax_inset.set_xlim(-2, 2)
ax_inset.set_ylim(-3, 3)
ax_inset.set_xticks([])
ax_inset.set_yticks([])

# Configurar leyenda y etiquetas de los ejes
ax.legend()

# Mostrar el gráfico
plt.show()
```


## <b>Time Series: Recrea el siguiente gráfico</b>

(Pista: Convierte la columna `last_review` a formato `DateTime`)

```python
abnb['last_review'] = pd.to_datetime(abnb['last_review'],format='%Y-%m-%d')


print(abnb['last_review'].dtype)



fig, ax = plt.subplots(figsize=(12, 8))

# Filtrar los datos para el rango de años 2016 a 2019
start_date = pd.to_datetime('2016-10')
end_date = pd.to_datetime('2019-04')

abnb_filtered = abnb.loc[(abnb['last_review'] >= start_date) & (abnb['last_review'] <= end_date)]



plt.plot(abnb_filtered['last_review'], abnb_filtered['price'],label='Price',color='orange')
plt.plot(abnb_filtered['last_review'],abnb_filtered['number_of_reviews'],label='Number of reviews')
plt.xlabel('Last Review')
plt.title('Fluctuations in Number of Reviews and Price over time',fontsize=20)
plt.legend(loc='upper left')
plt.show()
```
