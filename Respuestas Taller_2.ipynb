# Pandas 02 **Indexación, segmentación y creación de subconjuntos a partir de DataFrames en Python**

# 1) Rangos

```python
# a) Nos muestra la primera fila
print(surveys_df[0:1])
```

```python
# a) Le pedimos desde la primera fila, hasta el elemento número 4, en este caso hasta la fila 3 (si partimos a contar desde la fila 0 )
print(surveys_df[:4])
```

```python
# a) Le indicamos como argumento en el atributo que nos seleccione hasta la última fila, si agregamos -2 sería la penúltima y así 
#sucesivamente.
print(surveys_df[:-1])
```

```python
#b) en el caso de loc e iloc, tratamos de seleccionar una cantidad de filas y columnas especificas, sin embargo 
#la forma de ubicar las columnas y filas en ambos atributos es distinta, una es con el string de la columna y la otra con indices enteros.
print(surveys_df.iloc[0:4, 1:4])
```

# 3) Consultas

```python
# a) Se encuentran 55 filas con las dos condiciones indicadas 
df_sub = surveys_df[(surveys_df['year'] >= 1999) & (surveys_df['weight'] <= 8)]
#print(df_sub)
print(len(df_sub))
```

```python
# b) Crear lista de valores únicos en la columna "plot_id" que contienen la especie en particular

nombre_de_la_especie='NL' #Especie ejemplo
unique_plots = surveys_df[surveys_df['species_id'] == nombre_de_la_especie]['plot_id'].unique()

# Usar la función isin para seleccionar las filas que contienen esos valores
selected_rows = surveys_df[surveys_df['plot_id'].isin(unique_plots)]

# Contar el número de registros seleccionados
num_records = len(selected_rows)
print(num_records)
```

```python
# c) Usamos condicionales dentro de la ubicación por método de indices
df_temp=surveys_df[surveys_df['weight'] >= 0]
print(df_temp)
```

```python
#ocupamos el operador & para concatenar consultas
df_temp = surveys_df[(surveys_df['sex'] != 'F') & (surveys_df['sex'] != 'M')]
print(df_temp)
```

# 4) Revisando todo lo aprendido

```python
#a) concatenamos las consultas con el simbolo &
df = surveys_df[(surveys_df['sex'] != 'F') & (surveys_df['sex'] != 'M')]
```

```python
# b) 
df = surveys_df[(surveys_df['sex'] == 'F') | (surveys_df['sex'] == 'M') & (surveys_df['weight'] > 0)]  

print(df)
```

 # Pandas 04 **Combinando DataFrames con Pandas**

```python
pwd2='/home/mgcl/Escritorio/Quinto semestre/Taller DCD 3/Tareas/Taller 2/pandas_data/pandas_data/species.csv'
species_df = pd.read_csv(pwd2,sep=',')
```

```python
print(species_df.head())
```

# 1)

```python
survey_species = pd.merge(surveys_df, species_df, on='species_id')
print(survey_species)
```

```python
#Con el código groupby, agrupamos por parcela o sector donde se toma la muestra y con el atributo .nunique()
# contamos todos los valores únicos en la columna 'taxa'
taxa_por_parcela = survey_species.groupby(['plot_id'])['taxa'].nunique()

taxa_por_parcela
```

```python
# b) Un poco de los mismo que el código anterior, en este caso agrupamos por plot_id (parcela) y sex

taxa_por_sexo_parcela = survey_species.groupby(['plot_id', 'sex'])['taxa'].nunique()
taxa_por_sexo_parcela
```
